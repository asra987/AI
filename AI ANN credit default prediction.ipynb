#  4  label encoding and standardization

education is already label encoded 
only age and sex needs to be onehot encoded

```python
from sklearn.preprocessing import OneHotEncoder
ohe= OneHotEncoder()
onehot= ohe.fit_transform(dataset[["SEX", "MARRIAGE", "EDUCATION"]])
```

```python
onehot=onehot.toarray()
```

```python
x=pd.DataFrame(onehot)
```

```python
x.columns=['male','female', 'married', 'single', 'others status','graduate school','university', 'high school',"others education"]
x.head()
```

```python
dataset=dataset.reset_index(drop=True)
df= pd.concat([dataset, x],1)
df=df.drop(['SEX', 'MARRIAGE', 'EDUCATION','ID'],1)
df.head()
df.shape
# df.isnull().sum()
```

```python
X=df.drop(['target'],1)
X.head(5)
y=pd.DataFrame(df['target'])
y.columns=['target']
#y.head()
```

```python
X.head(2)
```

```python
x_num=X.drop(['PAY_0','PAY_2','PAY_3','PAY_4','PAY_5', 'PAY_6','male','female', 'married', 'single', 'others status','graduate school','university', 'high school',"others education"],1)
X1= X[['PAY_0','PAY_2','PAY_3','PAY_4','PAY_5', 'PAY_6','male','female', 'married', 'single', 'others status','graduate school','university', 'high school',"others education"]]
colsx=x_num.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
X_num=scaler.fit_transform(x_num)
X_num=pd.DataFrame(X_num)
X_num.columns=colsx
```

```python
X_num.head(2)
```

```python
X1.head(2)
```

```python
X_num=X_num.reset_index(drop=True)
X_f= pd.concat([X_num, X1],1)
```

```python
X_f.head()
```

```python
# from sklearn.cross_validation import train_test_split
# from sklearn.linear_model import LogisticRegression
# from sklearn.model_selection import cross_val_score
```

without feature selection acurae was 64

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_f, y, test_size = 0.30, random_state = 52)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import keras
```

```python
#sequential module is required to initialise our neural network
#dense module is required to build the layers of our ANN 
from keras.models import Sequential
from keras.layers import Dense
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
classifier= Sequential()
```

```python
#input layer and first hidden layer
classifier.add(Dense(output_dim = 15,init= 'uniform', activation='relu', input_dim= 29)) # number of nodes in input layer=29, number of nodes in output layer=1 [(29+1)/2=15]
#15 nodes in the hidden layer
```

```python
#second hidden layer
classifier.add(Dense(output_dim = 15,init= 'uniform', activation='relu')) #15=mean of nodes in input andoutput
```

```python
#output layer
classifier.add(Dense(output_dim = 1,init= 'uniform', activation='sigmoid')) 
```

```python
#compiling ANN
classifier.compile(optimizer= 'adam' , loss= 'binary_crossentropy' , metrics= ['accuracy']) 
#if binary outcome(logarithmic loss function)= binary_crossentropy
#if more than 2 categories in outcome= categorical_crossentropy
```

```python
#fitting ANN to traaining set
classifier.fit(X_train,y_train,batch_size = 10, epochs = 200)
```

```python
y_pred= classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
